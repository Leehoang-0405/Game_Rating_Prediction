```python
def categorize_year(year):
    if 1970 <= year < 1980:
        return '1970s'
    elif 1980 <= year < 1990:
        return '1980s'
    elif 1990 <= year < 2000:
        return '1990s'
    elif 2000 <= year < 2010:
        return '2000s'
    elif 2010 <= year < 2020:
        return '2010s'
    elif year >= 2020:
        return '2020s and Beyond'
    else:
        return 'Unknown'

ratings['Year_Category'] = ratings['Released_Year'].apply(categorize_year)

print(ratings.head())
```

```python
def create_histograms(data, columns=None, exclude=None):
    """
    Create customized histograms for numerical columns in the given DataFrame using seaborn.

    Parameters:
    data (pd.DataFrame): The DataFrame containing the data to plot.
    columns (list, optional): List of column names to plot. Defaults to None.
    exclude (list, optional): List of column names to exclude from plotting. Defaults to None.
    """

    # Set the overall aesthetic style of the plots
    sns.set(style="whitegrid")

    # If columns parameter is not provided, select all numerical columns
    if columns is None:
        columns = data.select_dtypes(include='number').columns

    # Exclude specific columns if provided
    if exclude is not None:
        columns = [col for col in columns if col not in exclude]

    # Define custom settings for each column type
    settings = {
        'Game_Type': {'color': 'green', 'bins': 10, 'kde': True},
        'Age_Category': {'color': 'green', 'bins': 15, 'kde': True},
        'Min_Players': {'color': 'green', 'bins': 10, 'kde': True},
        'Max_Players': {'color': 'green', 'bins': 10, 'kde': True},
        'Average_Complexity': {'color': 'green', 'bins': 25, 'kde': True},
        'Complexity_Raters': {'color': 'green', 'bins': 20, 'kde': True},
        'Average_Play_Time': {'color': 'green', 'bins': 30, 'kde': True},
        'Owner_Number': {'color': 'green', 'bins': 20, 'kde': True},
        'Trader_Number': {'color': 'green', 'bins': 20, 'kde': True},
        'HighInterest_Number': {'color': 'green', 'bins': 20, 'kde': True},
        'Interest_Number': {'color': 'green', 'bins': 20, 'kde': True},
        'Rater_Number': {'color': 'green', 'bins': 20, 'kde': True},
        'Comment_Number': {'color': 'green', 'bins': 20, 'kde': True},
        'Average_Rating': {'color': 'green', 'bins': 30, 'kde': True}
    }

    # Loop through each column in the DataFrame
    for column in columns:
        plt.figure(figsize=(10, 6))  # Create a new figure for each histogram

        # Determine settings based on column name
        column_settings = settings.get(column, {'color': 'blue', 'bins': 30, 'kde': True})

        # Create histogram using seaborn
        sns.histplot(data[column], kde=column_settings['kde'], bins=column_settings['bins'],
                     color=column_settings['color'], edgecolor='black')

        # Add title and labels
        plt.title(f'Histogram of {column}', fontsize=15)
        plt.xlabel(column, fontsize=12)
        plt.ylabel('Frequency', fontsize=12)

        # Display the histogram
        plt.show()

        # Ensure there is no buffering issue
        plt.close()

# Example usage
# Ensure you have the 'ratings' DataFrame with appropriate numerical columns
create_histograms(ratings, exclude=['Game_Type_N'])
```

```python
def create_boxplots(data, columns=None, exclude=None):
    """
    Create horizontal boxplots for numerical columns in the given DataFrame using seaborn.

    Parameters:
    data (pd.DataFrame): The DataFrame containing the data to plot.
    columns (list, optional): List of column names to plot. Defaults to None.
    exclude (list, optional): List of column names to exclude from plotting. Defaults to None.
    """
    sns.set(style="whitegrid")

    # If columns parameter is not provided, select all numerical columns
    if columns is None:
        columns = data.select_dtypes(include='number').columns

    # Exclude specific columns if provided
    if exclude is not None:
        columns = [col for col in columns if col not in exclude]

    # Create boxplots for each specified column
    for column in columns:
        plt.figure(figsize=(10, 6))  # Create a new figure for each boxplot

        # Create boxplot using seaborn
        sns.boxplot(x=data[column], color='blue')

        # Add title and labels
        plt.title(f'Boxplot of {column}', fontsize=15)
        plt.xlabel(column, fontsize=12)
        plt.ylabel('Value', fontsize=12)

        # Display the boxplot
        plt.show()

        # Ensure there is no buffering issue
        plt.close()

# Example usage
create_boxplots(ratings, exclude=['Game_Type_N'])
```

# Multi-Linear Regression

### **Select Features**

```python
features=['Interest_Number', 'HighInterest_Number','Average_Play_Time','Average_Complexity', 'Game_Type_N']
X=ratings[features]
y=ratings['Average_Rating']
y.head()
```

### **Split Data**

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2023)
print(X_train.head())
print(y_train.head())

print('Training dataset size:',X_train.shape)
print('Test dataset size:',X_test.shape)
```

### **Train a Linear Model**

```python
reg = linear_model.LinearRegression()
```

```python
reg=reg.fit(X_train, y_train)
```

### **Make Predictions**

```python
y_pred = reg.predict(X_test)
```

```python
plt.figure(figsize=(10, 8))

plt.scatter(y_test, y_pred, alpha=0.6, edgecolors='k', color='blue', s=60, label='Predicted Ratings')

min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Perfect Prediction')

plt.xlabel("Actual Average Ratings")
plt.ylabel("Predicted Average Ratings")
plt.title("Actual Average Ratings vs Predicted Average Ratings")

plt.legend()

plt.grid(True)

plt.show()
```

```python
#set the formatting for floating numbers
pd.set_option('display.float_format', lambda x: '%.0f' % x)

# Extract the 'Average_Play_Time' column
area = X_test['Average_Play_Time']

# Create the inspection DataFrame
inspection = pd.DataFrame({'Average_Play_Time': area, 'Actual': y_test, 'Predicted': y_pred})

# Display the first 20 rows
inspection.head(20)
```

### **Performance Evaluation**

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2:.2f}")
```

```python
from sklearn.metrics import mean_absolute_error

# Calculate and print the mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
from sklearn.metrics import mean_squared_error

# Calculate and print the root mean square error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Root Mean Square Error: {rmse:.2f}")
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Ensure all data in X is numeric
X = X.apply(pd.to_numeric, errors='coerce')
```

```python
# Convert boolean columns to integers
for col in X.columns:
    if X[col].dtype == 'bool':
        X[col] = X[col].astype(int)
```

```python
# Add intercept for VIF calculation
X_vif = X.copy()
X_vif['intercept'] = 1

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["variables"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

# Print VIFs to two decimal places
for index, row in vif_data.iterrows():
    print(f'{row["variables"]}: {row["VIF"]:.2f}')
```
